# Matplotlib Library
## Is a plotting library

```python
import matplotlib.pyplot as plt

%matplotlib inline # if this line included , then no need to write plt.show again and again
```

```python
import numpy as np
```

```python
## Simple Exemples
x=np.arange(0,10)
y=np.arange(11,21)
```

```python
##plotting using matplotlib

## plt scatter
plt.scatter(x,y,c='b') ## c means colour
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('1st Graph')
plt.savefig('1st Graph.png') ## To save the graph in image file
## Show Fighure
plt.show()
```

```python
y=x*x
```

```python
##plt plot
plt.plot(x,y,'ro--',linewidth=3,markersize='10') # 'r' signifies red colour # Then o-- means the pattern of the dotted file
## Show Fighure
plt.show()
```

```python
## Creating Subplots
plt.subplot(2,2,1)
plt.plot(x,y,'r--')
plt.subplot(2,2,2)
plt.plot(x,y,'bo-')
plt.subplot(2,2,3)
plt.plot(x,y,'g*-')
## Show Fighure
plt.show()
```

```python
x=np.arange(1,11)
y=3*x+5
plt.title('Function Example')
plt.xlabel('X axis')
plt.ylabel('Y axis')
plt.plot(x,y,'g')
## Show Fighure
plt.show()
```

```python
np.pi #Return value of pi
```

```python
## Compute the datas of X & Y-axis
x=np.arange(0,8*np.pi,0.1)
y=np.sin(x)
#Plot the graph
plt.plot(x,y)
## Show Fighure
plt.show()
```

```python
## Compute the data of X & Y-axis
x=np.arange(0,5*np.pi,0.1)
y_sin=np.sin(x)
y_cos=np.cos(x)

## Draw subplot 1 (Sin Graph)
plt.subplot(2,1,1)
plt.plot(x,y_sin,'b')
plt.title('Sine Graph')
## Draw Subplot2(Cos Graph)
plt.subplot(2,1,2)
plt.plot(x,y_cos,'r')
plt.title('Cos Graph')

## Show Fighure
plt.show()
```

```python
## Bar Plot
x=[2,8,10]
y=[11,16,9]
x2=[3,9,11]
y2=[6,15,7]
plt.bar(x,y)
plt.bar(x2,y2,color='g')
plt.title('Bar Graph')
plt.xlabel('X Axis')
plt.ylabel('Y Axis')
## Show Figure
plt.show()
```

# Box Plot using MAtplotlib

```python
data=[np.random.normal(0,std,100) for std in range(1,4)]
```

```python
data
```

```python
##Reactungular Box Plot
plt.boxplot(data,vert=True,patch_artist=True)
plt.show()
```

# Pie Chart

```python
# Data to Chart
labels_data='Python','C++','Java','Android'
sizes=[215,130,245,210]
colors_data=['gold','yellowgreen','lightcoral','lightskyblue']
explode_data=(0.1,0,0,0)
#Plot
plt.pie(sizes,explode=explode_data,labels=labels_data,colors=colors_data,autopct='%1.1f%%',shadow=True)
plt.axis('equal')
plt.show()
```
